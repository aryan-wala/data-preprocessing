```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../AB_NYC_2019.csv")
```

```python
df.head()
```

# Categorical Data

## Bar Graph

```python
sns.countplot(data = df, x="neighbourhood_group")
```

```python
sns.countplot(data = df, x="room_type")
```

```python
sns.countplot(data = df, x="neighbourhood_group",hue="room_type")
```

## Pie Chart
used when we want percentage values, eg how many % is the private covering in room_type 

```python
df["room_type"].value_counts().plot(kind = "pie", autopct="%.2f")
```

```python
df["neighbourhood_group"].value_counts().plot(kind = "pie", autopct="%.2f")
```

```python
df2 = sns.load_dataset("titanic")
```

```python
df2.head()
```

```python
sns.countplot(data = df2, x = "sex", hue="survived")
```

```python
df2["class"].value_counts().plot(kind = "pie", autopct = "%.2f")
```

# Numerical Data

## Histogram

```python
sns.histplot(data = df, x = "price")
```

```python
sns.histplot(data = df[df["price"]<500], x = "price")
```

```python
sns.histplot(data = df, x = "availability_365")
```

## Distribution
gives probability distribution

```python
sns.distplot(df["price"])
```

```python
sns.distplot(df["availability_365"])
```

## Boxplot

```python
sns.boxplot(df["price"])
```

```python
sns.boxplot(df["availability_365"])
```
