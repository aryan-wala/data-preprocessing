# Categorical+Numerical Data Analysis

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
sns.barplot(data=df, x="class", y ="age")
```

```python
#using estimator in plot
sns.barplot(data = df, x = "class", y = "fare", estimator = "sum")
```

```python
sns.barplot(data=df, x="class", y ="age", hue="sex") #hue is used to further categorise the x-axis category
```

```python
sns.boxplot(data=df, x="class", y ="age")
```

```python
sns.boxplot(data=df, x="class", y ="age", hue="sex")
```

```python
sns.violinplot(data=df, x="class", y ="age")
```

```python
sns.distplot(df[df["class"] == "Second"]["age"])
```

# Only Categorical Data
heat map is used when there are more categories

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
sns.countplot(data = df, x = "survived", hue = "class")
```

```python
x = pd.crosstab(df["survived"],df["class"])
```

```python
x
```

```python
sns.heatmap(x,annot=True, fmt = ".1f", cmap="coolwarm")
```

# Numerical Data

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

## Scatterplot

```python
sns.scatterplot(data = df, x = "age", y = "fare")
```

```python
sns.scatterplot(data = df, x = "age", y = "fare", hue="survived")
```

```python
df2 = sns.load_dataset('tips')
df2.head()
```

```python
sns.scatterplot(df2, x = "total_bill", y= "tip")
```

```python
sns.scatterplot(df2, x = "total_bill", y= "tip", hue="sex")
```

```python
sns.scatterplot(df2, x = "total_bill", y= "tip", hue="sex", style = "smoker")
```

## Joint Plot

```python
sns.jointplot(df2, x = "total_bill", y= "tip")
```

```python
sns.jointplot(df2, x = "total_bill", y= "tip", hue="sex")
```

```python
sns.jointplot(df2, x = "total_bill", y= "tip", kind = "reg")
```

```python
df2.corr(numeric_only = True)
```
